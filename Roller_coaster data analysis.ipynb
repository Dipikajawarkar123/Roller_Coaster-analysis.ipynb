```python
df=df.iloc[ :, [0,3,4,7,41,42,43,44,45,50,53,54,55]]
```

```python
df.dtypes
```

```python
#renaming columns name 
df=df.rename(columns={'coaster_name':'Coaster_Name',
                     'year_introduced':'Year_Introduced',
                     'opening_date_clean':'Opening_Date',
                     'speed_mph':'Speed_mph',
                     "height_ft":"Height_ft",
                     "Inversions_clean":"Inversions",
                     "Gforce_clean":"Gforce"})
```

```python
df.columns
```

```python
# checking null values

df.isnull().sum()
```

```python
df.duplicated()
```

```python
df=df.loc[~df.duplicated(subset=['Coaster_Name','Location','Opening_Date'])]\
    .reset_index(drop=True).copy()
```

```python
df.shape
```

```python
ax=df['Year_Introduced'].value_counts()\
    .head(10) \
    .plot(kind='bar')
plt.title('Top 10 Coaster Introduced')
ax.set_xlabel('Year Introduced')
ax.set_ylabel('Count')
 
```

```python
ax = df['Speed_mph'].plot(kind='hist', bins=20, title = 'Coaster Speed (mph)')

ax.set_xlabel('Speed (mph)')
```

```python
ax = df['Speed_mph'].plot(kind='kde', title = 'Coaster Speed (mph)')

ax.set_xlabel('Speed (mph)')

plt.grid()
```

```python
df['Location'].value_counts()
```

```python
df_query=df.query('Location != "Other" ' ).groupby('Location')['Speed_mph'] .agg(['mean','count']) \
    
```

```python
query=df_query.query('count>=10').sort_values('mean')['mean']
```

```python
plt.figure(figsize=(12,5))
query.plot(kind='barh')
plt.title('Average Coaster Speed By Location')
plt.ylabel('Location')
plt.xlabel('Average Coaster Speed')
```
